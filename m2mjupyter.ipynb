## Transactions per second, per instance

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as po
df = pd.read_csv('https://raw.githubusercontent.com/unpetee/getocidisks/master/tps.csv')
df['DATETIME']= pd.to_datetime(df['DATETIME'],format="%d/%m/%Y %H:%M")
df.loc[df.DATA > 0, 'B'] = df.DATA
fig = px.line(df, x = 'DATETIME', y = 'B',color='INSTANCE', title='TPS x INSTANCE')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
df['DATETIME']= pd.to_datetime(df['DATETIME'],format="%d/%m/%Y %H:%M").dt.strftime('%d/%m/%Y %H')
df['DATETIME']= pd.to_datetime(df['DATETIME'],format="%d/%m/%Y %H")
df.loc[df.DATA > 0, 'B'] = df.DATA
gb_date = df.groupby("DATETIME")["B"].sum()
fig2 = px.line(gb_date,title='TPS TOTAL')
fig2.show()
```

## TOP SQLs - LAST day.

```python
import pandas as pd
import datetime as dt
import plotly.express as px
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/unpetee/getocidisks/master/topSQLs.csv')
df['DATETIME']= pd.to_datetime(df['DATETIME'],format="%d/%m/%Y %H:%M")
start = dt.datetime.strptime('2021-10-26 00:00:00', '%Y-%m-%d %H:%M:%S')
end = dt.datetime.strptime('2021-10-26 16:59:59', '%Y-%m-%d %H:%M:%S')
sdf = (df.DATETIME >= start) & (df.DATETIME <= end)
df.loc[sdf,'DATE'] = df.DATETIME
#df['miSQLID'] = np.where(np.isnat(df.DATE), '0',df.SQLID)
#df = df[df['miSQLID'] > 'a']

df = df[df['DATE'] >= start]
df = df[df['DATE'] <= end]
df = df[df['INSTANCE'] == 1]
#df
fig = px.line(df, x = 'DATE', y = 'CPUTIME',color='SQLID', title='TOP5 SQL<b> INSTANCE 1<b>')
fig.update_xaxes(rangeslider_visible=True)
fig.show()

######

df = pd.read_csv('https://raw.githubusercontent.com/unpetee/getocidisks/master/topSQLs.csv')
df['DATETIME']= pd.to_datetime(df['DATETIME'],format="%d/%m/%Y %H:%M")
start = dt.datetime.strptime('2021-10-26 00:00:00', '%Y-%m-%d %H:%M:%S')
end = dt.datetime.strptime('2021-10-26 16:59:59', '%Y-%m-%d %H:%M:%S')
sdf = (df.DATETIME >= start) & (df.DATETIME <= end)
df.loc[sdf,'DATE'] = df.DATETIME

df = df[df['DATE'] >= start]
df = df[df['DATE'] <= end]
df = df[df['INSTANCE'] == 2]
#df
fig = px.line(df, x = 'DATE', y = 'CPUTIME',color='SQLID', title='TOP5 SQL<b> INSTANCE 2<b>')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
